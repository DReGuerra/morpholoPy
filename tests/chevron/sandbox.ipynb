```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import feature, io, util

# aray size NxN
N = 900             # pixels
# scale
L = N               # um
x_scale = L/N       # um/pixel
# create NxN array
img = np.zeros((N,N))
img2 = np.zeros((N,N))

w = 50
for i in np.arange(0,N,w*2):
    img[:, i:i+w] = 255
    img2[:, i:i+w] = 255

# make chevron image from the vertical lines image
period = N/20
r = 1
for i in np.arange(1,N-1,1):
    if r > 0:
        tmp = img2[i-1, :]
        tmp = np.delete(tmp, -1)
        tmp = np.insert(tmp, 0, 0)
        img2[i, :] = tmp
        if i%period == 0: r = r*-1
    else:
        tmp = img2[i-1, :]
        tmp = np.delete(tmp, 0)
        tmp = np.insert(tmp, -1, 0)
        img2[i, :] = tmp
        if i%period == 0: r = r*-1

plt.imshow(img2, cmap='gray')
```
